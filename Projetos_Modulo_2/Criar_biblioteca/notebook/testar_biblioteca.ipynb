# Imports

```python
from image_preprocessing.preprocessing import tools, pipeline
from image_preprocessing.utils import io, display_images
```

# Leitura e preprocessamento das imagens

```python
image_path = "../data/raw/cat.jpg"

image = io.load_image(image_path,grayscale=False) 

reshape_img = tools.reshape_image(image)

scale_img = tools.scale_image(image)

aug_img = tools.random_augumentation(image) 

preprocess_img = pipeline.preprocess_pipeline(image) 
```

# Plotando resultado das imagens resultantes

```python
imagemtemp = display_images.plot_images(image,reshape_img,scale_img,aug_img,preprocess_img)
imagemtemp
```

# Salvando imagens

```python
io.save_image(reshape_img,'../data/preprocessed/reshaped_image.jpg')
io.save_image(scale_img,'../data/preprocessed/scaled_image.jpg')
io.save_image(aug_img,'../data/preprocessed/augumented_image.jpg')
io.save_image(preprocess_img,'../data/preprocessed/preprocessed_image.jpg')
```
